### Feature selection 

#### Feature selection - Variance threshold

```python
# Make all variables be on the same scale in order to look at VarianceThresholds (use MinMaxScaler)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(numerical)
numerical_scaled = scaler.transform(numerical)
```

```python
numerical_scaled_df = pd.DataFrame(numerical_scaled, columns = numerical.columns)
numerical_scaled_df.head()
```

```python
#if the variance of a column is lower than the threshold, the column will be dropped
from sklearn.feature_selection import VarianceThreshold

var_threshold = 0.08
sel = VarianceThreshold(threshold = var_threshold)

sel = sel.fit(numerical_scaled_df)
temp = sel.transform(numerical_scaled_df)
```

```python
temp_df = pd.DataFrame(temp)
print(numerical_scaled_df.shape)
print(temp_df.shape)
temp_df.head()
```

```python
# the following columns are kept
sel.get_feature_names_out()
```

```python
temp_df.columns = sel.get_feature_names_out()
temp_df.head()
```

## RF - classification (when y ='TARGET_B')

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import make_scorer, accuracy_score,confusion_matrix,ConfusionMatrixDisplay,precision_score, recall_score
```

```python
X = data_new.drop(['TARGET_B','TARGET_D'],axis=1)
y = data_new['TARGET_B']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
data_new['TARGET_B'].value_counts()
```

### downsampling

```python
#since the number of 0 is way more than 1, we need to do downsampling first
trainset = pd.concat([X_train,y_train], axis =1)
```

```python
category_0 = trainset[trainset['TARGET_B']==0].sample(len(trainset[trainset['TARGET_B']==1]))
category_1 = trainset[trainset['TARGET_B']== 1 ]
```

```python
trainset_new = pd.concat([category_0, category_1], axis = 0)
trainset_new = trainset_new.sample(frac =1) #randomize the rows
```

```python
trainset_new['TARGET_B'].value_counts()
```

```python
X_train = trainset_new.drop(['TARGET_B'],axis=1)
y_train = trainset_new['TARGET_B']
print(X_train.shape)
print(y_train.shape)
```

### modelling

```python
# Define a custom scorer using the accuracy metric
scorer = make_scorer(accuracy_score)
```

```python
max_depth_choices= [3,5,7] # A list of the possible values of max_depth to try
n_estimators = [30,100,300]      
```

```python
# Create the  grid 
grid = {'max_depth': max_depth_choices,
       'n_estimators':n_estimators}
```

```python
RF_clf = RandomForestClassifier()
```

```python
#Create a GridSearchCV object
RF_clf_grid = GridSearchCV(estimator = RF_clf, param_grid = grid, cv = 5, scoring = scorer)
```

```python
# Fit the GridSearchCV object to the data
RF_clf_grid.fit(X_train, y_train)

print("Best parameters:", RF_clf_grid.best_params_)
print("Best score:", RF_clf_grid.best_score_)
```

```python
print("The accuracy for the Random Forest classification in the TRAIN set is {:.2f}".format( RF_clf_grid.score(X_train, y_train)))
print("The accuracy for the Random Forest classification in the TEST  set is {:.2f}".format( RF_clf_grid.score(X_test, y_test)))
```

```python
y_test_pred = RF_clf_grid.predict(X_test)
display(y_test.value_counts())
display(confusion_matrix(y_test, y_test_pred))
```

```python
cm_test = confusion_matrix(y_test, y_test_pred)
import matplotlib.pyplot as plt
disp = ConfusionMatrixDisplay(cm_test,display_labels=RF_clf_grid.classes_)
disp.plot()
plt.show()
```

```python
#another error metric : cohen_kappa_score
scorer2 = make_scorer( precision_score)
RF_clf_grid2 = GridSearchCV(estimator = RF_clf, param_grid = grid, cv = 5, scoring = scorer2)
RF_clf_grid2.fit(X_train, y_train)

print("The precision_score for the Random Forest classification in the TRAIN set is {:.2f}".format( RF_clf_grid2.score(X_train, y_train)))
print("The precision_score for the Random Forest classification in the TEST  set is {:.2f}".format( RF_clf_grid2.score(X_test, y_test)))
```

```python
y_test_pred = RF_clf_grid2.predict(X_test)
precision_score(y_test, y_test_pred)
```
